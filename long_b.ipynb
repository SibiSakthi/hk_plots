```python
def plot_beacon(Y, folder, filtered_df,X='time_stamp'):
    try:
        if filtered_df is None or filtered_df.empty:
            print("DataFrame is empty or None.")
            return
        try:
            if X in filtered_df.columns and X != 'time_stamp':
                filtered_df[X] = filtered_df[X].astype(float)
                
            if Y in filtered_df.columns:
                filtered_df[Y] = filtered_df[Y].astype(float)
            else:
                print(f"Column {Y} does not exist in DataFrame.")
        except Exception as e:
            print(f"Error occurred: {str(e)}")
        x = filtered_df[X]
        y = filtered_df[Y]
        plt.figure(figsize=(16, 10))#adjust size
        plt.plot(x, y, marker='o', linestyle='-', color='b', label=f'{X} vs {Y} (CRC=True)')
        plt.title(f'{X} vs {Y} (CRC=True)')
        plt.xlabel(X)
        plt.ylabel(Y)
        plt.legend()
        plt.grid(True)
        plt.xticks(rotation='vertical')
        plt.tight_layout()
        save_dir = folder
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)
        filename = os.path.join(save_dir, f'{Y+" vs "+X}_plot.png')
        plt.savefig(filename)
        print(folder,"------------+==========+-----------")
        plt.show()
    except:
        print(f"Error occurred: {str(e)}")   
```
